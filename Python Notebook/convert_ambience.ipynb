```python
import pandas as pd
```

```python
df=pd.read_csv('Business_Attributes_Final.csv')
df.head()
```

```python
df_flt = df[df['attributes.Ambience'].notna()]
```

```python
df_flt['attributes.Ambience'] = df_flt['attributes.Ambience'].apply(lambda x: eval(x))
```

```python
data = [df_flt['attributes.Ambience'], df_flt['business_id']]

headers = ["attributes_Ambience", "business_id"]

df3 = pd.concat(data, axis=1, keys=headers)
```

```python
type(df3['business_id'][0])

df3.shape
```

```python
df3.dropna(inplace=True)
```

```python
out = df3.to_json(orient='records')[1:-1].replace('},{', '} {')
```

```python
import json
count = 0
final_data = []
for index, row in df3.iterrows():
    temp_dict = row['attributes_Ambience']
    temp_dict['business_id_orig'] = row['business_id']
    final_data.append(temp_dict)

with open('Business_Ambience_Attributes.json', 'a') as outfile:
    json.dump(final_data, outfile, indent=2)
```
